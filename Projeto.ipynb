# ***Analisei as tendências das imigrações do Brasil ao longo dos anos***



```python
df.set_index('País', inplace=True)
```

```python
#criei uma variável para ter focado só os anos
anos = list(map(str, range(1980, 2014)))
print(anos)
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

Agora vou começar a manipular esses dados obtidos após o tratamento, para fazer os gráficos para melhor compreeensão  

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title("Imigração do Brasil para o Canadá ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Número de imigrantes")
plt.show()
```

# ***Nessa parte farei o mesmo com a Argentina para fazer a comparação dela com o Brasil***

```python
argentina = df.loc['Argentina', anos]
```

```python
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes' : argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```python
plt.figure(figsize=(8, 9 ))
plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"])
plt.plot(dados_argentina["ano"], dados_argentina["imigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
plt.yticks([500, 1000, 1500, 2000, 2500])
plt.title("Comparação de Imigrantes da Argentina e Brasil ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Número de Imigrantes")
plt.show()

#O azul é o Brasil e o Laranja a Argentina

# Pergunta: Há alguma tendência ou padrão comum nos dados dos dois países?

# Resposta: Sim, há um padrão de crescimento do número de imigrantes ao longo do tempo em ambos os países, especialmente após 1990.
# Ambos apresentam oscilações com subidas e descidas em momentos semelhantes, indicando influências regionais ou globais nos fluxos migratórios.
# Notamos que a partir de 2000 ambos começam a crescer, mas o Brasil (linha azul) segue crescendo de forma mais intensa e consistente após 2005,
# enquanto a Argentina (linha laranja) diminui depois de atingir seu pico.

# Pergunta: Quais são os períodos com maior número de imigrantes nos dois países?

# Resposta: No Brasil, o maior número de imigrantes ocorre entre 2005 e 2010, chegando a mais de 2500 imigrantes, sendo o pico do gráfico.
# Na Argentina, o maior número ocorre entre 2003 e 2005, com cerca de 1700-1800 imigrantes. Antes disso, a Argentina teve picos menores em 1984 e 1994,
# mas de menor intensidade. Resumindo: os maiores picos aconteceram nos anos 2000, mas o Brasil manteve crescimento enquanto a Argentina teve queda.
```

# ***Criando figuras para Aprimorar e deixar masi detalhado o gráficos***

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil["ano"], dados_brasil["imigrantes"], lw=3, color='g')
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=16, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('Imigração_brasil_canada.png', transparent=False, dpi=800, bbox_inches='tight')

plt.show()
```

# ***analisei as tendências de imigração para os países da America do sul sendo os 4 maiores(Brasil, Colômbia, Peru e Argentina) o foco foi comparar as tendências em uma unica figura***

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n 1980 a 2013')


axs[0,0].plot(anos, df.loc['Brasil', anos])
axs[0,0].set_title("Brasil")

axs[0,1].plot(anos, df.loc["Colômbia", anos])
axs[0,1].set_title("Colômbia")

axs[1,0].plot(anos, df.loc["Argentina", anos])
axs[1,0].set_title("Argentina")

axs[1,1].plot(anos, df.loc["Peru", anos])
axs[1,1].set_title("Peru")

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

agora irei ver o Brasil com comparado a todos os países da América do Sul

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')


fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title ('América do Sul: Brasil foi o quarto país com mais imigrantes\n para o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
ax.grid(False)

fig.savefig('Imigração_america_sul.png', transparent=False, dpi=800, bbox_inches='tight')

plt.show()
```

```python
import plotly.graph_objs as go

fig = go.Figure()


fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes do Brasil', line=dict(width=4))
)


fig.add_trace(
    go.Scatter(x=[dados_argentina['ano'].iloc[0]], y=[dados_argentina['imigrantes'].iloc[0]], mode='lines', name='Imigrantes da Argentina', line=dict(width=4))
)


fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil e da Argentina para o Canadá no período de 1980 a 2013',
        x=0.1,

        font=dict(size=18)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600
)


frames = []
for i in range(len(dados_brasil)):
    frame_data = [
        go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1]),
        go.Scatter(x=dados_argentina['ano'].iloc[:i+1], y=dados_argentina['imigrantes'].iloc[:i+1])
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig.frames = frames


fig.show()
```
